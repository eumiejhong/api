```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```
np.random.seed(1221)
```

```
df = pd.DataFrame()
df['rand'] = np.random.rand(100)  # Add a column of random numbers between 0 and 1.
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2
```

```
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1, 2, 2)
plt.plot(df['rand_shift'], color='green')
plt.ylabel('Shifted Values')
plt.title('Shifted Series')
plt.show()
```

```
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1, 2, 2)
plt.scatter(x = df['rand_sq'], y = df['rand'], color='green')
plt.ylabel('Squared Values')
plt.title('Squared Series')

plt.tight_layout()
plt.show()
```

```
x = np.random.normal(10, 5, 1000)
```

```
plt.hist(x, bins=40, color='red')
plt.title('Default Bin Placement Demo')
plt.xlabel('Random Values')
plt.show()
```

```
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)
```

```
plt.hist(x, color='blue', bins=np.arange(-10, 40), alpha=.5) 
plt.hist(y, color='red', bins=np.arange(-10, 40), alpha=.5)
plt.title('Manually setting bin placement')
plt.xlabel('Random Values')

plt.show()
```

```
plt.hist(x, normed=True, color='blue', bins=np.arange(-10, 40), alpha=.5) 
plt.hist(y, normed=True, color='red', bins=np.arange(-10, 40), alpha=.5)
plt.title('Normed histograms')
plt.xlabel('Random Values')

plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
```

```
heart = pd.read_csv('https://raw.githubusercontent.com/eumiejhong/heartdata/master/heart.csv')
print(heart)    
```

```
plt.scatter(heart['age'], heart['thalach'])
plt.title("Age and Maximum Heart Rate Achieved")
plt.xlabel("Age (Years)")
plt.ylabel("Max Heart Rate")
plt.grid(True)
plt.show()

#The following is a scatterplot that examines age in years and maximum heart rate achieved after cardio.
#Overall there appears to be a weak linear negative relationship of age and maximum heart rate achieved.
```

```
plt.subplot(221)
plt.scatter(heart['age'], heart['thalach'])
plt.title("Age and Maxiumum Heart Rate Achieved")
plt.xlabel("Age in years")
plt.ylabel("Max Heart Rate")
plt.show()

plt.subplot(221)
plt.scatter(heart['trestbps'], heart['chol'])
plt.title("Resting Blood Pressure and Cholesterol Levels")
plt.xlabel("Blood Pressure")
plt.ylabel("Cholesterol Level")
plt.show()

#The following are subplots examining both age vs. maximum heart rate (after cardio) as well as blood pressure (resting rate) vs. cholesterol levels. 
#Overall there appears to be a weak linear negative relationship of age and maximum heart rate achieved.
#In terms of the relationship between blood pressure and cholesterol level, there appears to be no relationship between the two variables.
```

```
thal = heart['thal']
heart_rate = heart['thalach']
plt.hist([thal], color=['orange'])
plt.xlabel("Heart Disease Levels")
plt.ylabel("Patients")
plt.xticks(range(0, 4))
plt.yticks(range(1, 200, 10))
plt.title('Heart Disease Frequency in Patients')
plt.show()

#The following is a histogram plot of the different levels of heart disease found in patients.
#level 0  - 3 are levels considered "normal", with 3 being the maximum. Overall, the most common level is 2 among the patients.
```

```
heart.boxplot(column='chol', by='sex');

# 1 = male and 0 = female. The following are boxplots of cholesteral levels between male and female patients. 
#Overall, the range of cholesterol was wider and higher for men compared to women. 
#Their median cholesterol values were also higher compared to women.
```
